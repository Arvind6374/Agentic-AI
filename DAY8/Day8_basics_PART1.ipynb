```python

#What is a Document?
#A document is raw input data (text, PDF, webpage, etc.) stored with optional metadata.

from llama_index.core import Document

text1 = """
Python is a popular programming language used for AI, ML, and Data Science.
It is easy to learn and has a huge ecosystem.
"""

text2 = """
JavaScript is mainly used for web development.
It runs inside the browser and powers interactive web apps.
"""

doc1 = Document(text=text1)
doc2 = Document(text=text2)

docs = [doc1, doc2] # doc = [doc1,doc2,doc3,doc4]

print(type(doc1))

print("\ncombined documents\n")
docs
```

```python

#What is a Document?
#A document is raw input data (text, PDF, webpage, etc.) stored with optional metadata.

from llama_index.core import Document

text1 = """
Python is a popular programming language used for AI, ML, and Data Science.
It is easy to learn and has a huge ecosystem.
"""

text2 = """
JavaScript is mainly used for web development.
It runs inside the browser and powers interactive web apps.
"""

doc1 = Document(text=text1)
doc2 = Document(text=text2)

docs = [doc1, doc2] # doc = [doc1,doc2,doc3,doc4]

print(type(doc1))

print("\ncombined documents\n")
docs
```

```python
##Nodes — Chunked version of documents

## What is a Node?
## A Node is a small chunk of a document (LlamaIndex automatically splits text).

## Why?
## Smaller chunks = more accurate retrieval

from llama_index.core.node_parser import SimpleNodeParser

parser = SimpleNodeParser()
nodes = parser.get_nodes_from_documents(docs)
len(nodes)
nodes[1].text
```

```python
from llama_index.core import SimpleDirectoryReader

loader = SimpleDirectoryReader("./Data")
documents = loader.load_data()

print(len(documents))

print(documents[0].text)
```
